# 1.2

```python
#LCC
def LCC(G):
    Subg = [G.subgraph(c).copy() for c in nx.connected_components(G)]
    return [len(subgraph.nodes) for subgraph in Subg]
print(LCC(G))
```

```python
# average cluster coefficient
def avg_cluster_coefficient(G):
    c = nx.clustering(G)
    c = sorted(c.items(), key=lambda pair: pair[1], reverse=True)
    c_avg = sum([i[1] for i in c]) / len(c)
    return c_avg

print(nx.average_clustering(G))
print(f'avg clustering coefficient: {avg_cluster_coefficient(G)}')
```

## W BC

```python
# betweenness centrality
W_bet_centrality=nx.betweenness_centrality(G, normalized=False, weight='v_flows')
nx.set_node_attributes(G, W_bet_centrality, 'betweenness')

# To ataframe using the nodes as the index
df = pd.DataFrame(index=G.nodes())
# df['station_name'] = pd.Series(nx.get_node_attributes(G, 'station_name'))
df['Weighted Betweenness Centrality'] = pd.Series(nx.get_node_attributes(G, 'betweenness'))

W_BC_df_sorted = df.sort_values(["Weighted Betweenness Centrality"], ascending=False)
W_BC_df_sorted.head(10)
#W_BC_df_sorted.head(10).to_excel('./W_BC_df_sorted.xlsx')
```

```python

def avg_weighted_path_length(G):
    components = nx.connected_components(G)
    #find the largest component
    largest_component = max(components, key=len)
    # Create a subgraph
    Largest_subgraph = G.subgraph(largest_component)
    avg = nx.average_shortest_path_length(Largest_subgraph, weight='v_flows')
    return avg
# components = nx.connected_components(G)
# largest_component = max(components, key=len)
# Largest_subgraph = G.subgraph(largest_component)
# avg = nx.average_shortest_path_length(Largest_subgraph, weight='flows')
print(avg_weighted_path_length(G))
```
